```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
```

```python
for i in  data.columns:
    if data[i].dtypes == 'object':
        data [i] = ord_enc.fit_transform(data[i].values.reshape(-1,1))
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(24,9))
sns.heatmap(data.corr(),annot=True,linewidth=0.1, fmt ="0.2f")
```

```python
data.skew()
```

```python
plt.figure(figsize=(20,25),facecolor='grey')
plotnumber = 1
for column in data:
    if plotnumber <72:
        ax = plt.subplot(18,4,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(20,30))
graph = 1

for column in data:
    if graph<=72:
        plt.subplot(18,4,graph)
        ax=sns.boxplot(data[column])
        plt.xlabel(column,fontsize=15)
    graph+=1
plt.show()
```
